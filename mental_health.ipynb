```python
x=a.drop('Code',axis=1)
Entity={level: i for i, level in enumerate(sorted(set(x['Entity'])))}
x['Entity'] = x['Entity'].map(Entity)
```

```python
import matplotlib.pyplot as plt
for feature in x.columns:
    plt.figure(figsize=(8, 6))
    plt.hist(x[feature], bins=20, alpha=0.5)
    plt.title(f'Histogram of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()
```

```python
import numpy as np
for col in x:
    counts, bin_edges = np.histogram(x[col].dropna(), bins=20)
    x[col] = {"counts": counts, "bin_edges": bin_edges}
```

```python
import pickle
with open("histogram_model.pkl", "wb") as f:
    pickle.dump(x, f)
```
